# INTRODUCTION:

* The "Panchayat Finance Data 2020-2021" dataset provides financial data into various districts across different states in India. 
* It includes key financial metrics related to panchayat-level transactions, such as e-Governance System for Panchayati Raj Institutions (eGSPI), payment verification initiation, approval, and closure of financial transactions. 
* This dataset can be used to analyze financial trends, assess the efficiency of fund utilization, and identify gaps in financial governance at the grassroots level.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Panchayat Finance Data 2020-2021.csv")
df
```

# Dataset Description:

* finyear: The financial year for which the data is recorded.
* state_code: A numerical code representing the state.
* state_name: The name of the state.
* district_code: A numerical code representing the district.
* district_name: The name of the district.
* onboard_egspi: Indicates whether the district has onboarded the e-Gram Swaraj Public Information system (0 = No, 1 = Yes).
* pv_initiated: The number of payments (or payment verifications) initiated.
* pv_approved: The number of payments (or payment verifications) approved.
* year_closed: Indicates whether the financial year processes are closed (0 = No, 1 = Yes).

# Payement Verification Information

```python
df_grouped_state = df.groupby("state_name")["pv_initiated", "pv_approved", "year_closed"].sum()
display(df_grouped_state)
```

Most of the Payent verifications are initiated. Some of them are Completed.

# Data Visualisation through various plots

```python
plt.figure(figsize=(12,6))
sns.barplot(x=df_grouped_state.index, y=df_grouped_state["pv_initiated"], palette="viridis")
plt.xticks(rotation=90)
plt.title("Total Payment Vouchers Initiated by State")
plt.xlabel("State")
plt.ylabel("Count")
plt.show()
```

More number of vouchures are initiated by the uttar pradesh state. and the second highest is Madhya pradesh

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_grouped_state.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Financial Metrics")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="state_name", y="pv_initiated", data=df)
plt.xticks(rotation=90)
plt.title("Distribution of Payment Vouchers Initiated by State")
plt.xlabel("State")
plt.ylabel("Payment Vouchers Initiated")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x=df["pv_initiated"], y=df["pv_approved"], hue=df["state_name"], palette="tab10")
plt.title("Payment Vouchers Initiated vs Approved")
plt.xlabel("Payment Vouchers Initiated")
plt.ylabel("Payment Vouchers Approved")
plt.show()
```

Most of the Payment vochures got intiated than the approved.

```python
plt.figure(figsize=(12,6))
df_grouped_district = df.groupby("district_name")["pv_initiated", "pv_approved"].sum()
df_grouped_district.sort_values("pv_initiated", ascending=False).head(10).plot(kind="bar", figsize=(12,6))
plt.title("Top 10 Districts by Payment Vouchers Initiated")
plt.xlabel("District")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
top_5_states = df_grouped_state.sort_values("pv_approved", ascending=False).head(5)
display(top_5_states)
```

```python
df_grouped_state["approval_rate"] = (df_grouped_state["pv_approved"] / df_grouped_state["pv_initiated"]) * 100
display(df_grouped_state[["approval_rate"]].sort_values("approval_rate", ascending=False))
```

```python
plt.figure(figsize=(12,6))
df_grouped_state[["pv_initiated", "pv_approved"]].plot(kind="line", marker='o', figsize=(12,6))
plt.title("Initiated vs Approved Payment Vouchers by State")
plt.xlabel("State")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,10))
df_grouped_state["approval_rate"].plot(kind="pie", autopct="%1.1f%%", cmap="Set2")
plt.title("Approval Rate Distribution by State")
plt.ylabel("")
plt.show()
```

In the Madhya Pradesh state the Approval rate is high

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x="year_closed", y="pv_initiated", label="Initiated", marker='o')
sns.lineplot(data=df, x="year_closed", y="pv_approved", label="Approved", marker='s')
plt.title("Trend of Payment Vouchers Over the Years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
df_grouped_district["approval_rate"] = (df_grouped_district["pv_approved"] / df_grouped_district["pv_initiated"]) * 100
display(df_grouped_district[["approval_rate"]].sort_values("approval_rate", ascending=False))
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_grouped_district.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Initiated and Approved Vouchers at District Level")
plt.show()
```

```python
df["year_closed"] = pd.to_numeric(df["year_closed"], errors='coerce')  # Ensure it's numeric
df["month_closed"] = pd.to_datetime(df["year_closed"].astype(str), format='%Y', errors='coerce').dt.month

plt.figure(figsize=(12,6))
sns.lineplot(data=df, x="month_closed", y="pv_initiated", label="Initiated", marker='o')
sns.lineplot(data=df, x="month_closed", y="pv_approved", label="Approved", marker='s')
plt.title("Monthly Trend of Payment Vouchers")
plt.xlabel("Month")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
df["pv_approved"].plot(kind="hist", bins=20, alpha=0.7, color='blue', edgecolor='black')
plt.title("Distribution of Approved Payment Vouchers")
plt.xlabel("Number of Approved Vouchers")
plt.ylabel("Frequency")
plt.show()
```

# Accuracy

```python
total_initiated = df["pv_initiated"].sum()
total_approved = df["pv_approved"].sum()

approval_accuracy = (total_approved / total_initiated) * 100
print(f"Overall Approval Accuracy: {approval_accuracy:.2f}%")
```
